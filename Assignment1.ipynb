#### Assignment 1: Temperature Check 

```python
# Check current temperature
temperature = int(input('Enter current temperature: '))

if temperature < 10:
    print("It's too cold!")

elif temperature >= 10 and temperature <= 25:
    print("It's cool outside")

else:
    print("It's hot outside!")
```

#### Assignment 2: ATM Withdrawal System

```python
balance = int(input("Enter your current balance: "))
withdrawal = int(input("Enter your withdrawal amount: "))

if withdrawal > balance:
    print("Insufficient Balance")

else:
    print("Transaction Successful")
    print("Remaining Balance:", (balance-withdrawal))
```

#### Assignment 3: Student Grade Calculator 

```python
marks = int(input("Enter your marks (0-100): "))

if marks >= 90 and marks <= 100:
    print("your grade A+")

elif marks >= 80 and marks <= 89:
    print("your grade A")

elif marks >= 70 and marks <= 79:
    print("your grade B")

elif marks >= 60 and marks <= 69:
    print("your grade C")

else:
    print("your grade Fail")
    
```

#### Assignment 4: Bus Ticket Discount

```python
age = int(input("Enter your age: "))

if age < 5:
    print("Your ticket is Free")

elif age >= 5 and age <= 18:
    print("You will get 50% Discount")

elif age >=60:
    print("You will get 30% Discount")

else:
    print("Your ticket is Full price")
```

#### Assignment 5: Restaurant Bill with Discount

```python
bill = float(input("Enter your total bill amount: "))

if bill > 1000:
    print("Your bill amount:", (bill-bill*.01))

else:
    print("Your bill amount:", bill)
```

#### Assignment 7: Traffic Light System

```python
color = input("Enter your color: ")

if color == "red":
    print("Stop")

elif color == "yellow":
    print("Ready to go")

elif color == "green":
    print("Go")

else:
    print("Invalid color")
```

#### Assignment 8: Leap Year Checker

```python
year = int(input("Enter a year: "))

if year % 400 == 0:
    print("Leap Year")

elif year % 4 == 0 and year % 100 !=0:
    print("Leap Year")

else:
    print("Not a Leap Year")
```

#### Assignment 10: Electricity Bill Calculator 

```python
units = int(input("Enter your consumed units: "))

if units <= 100:
    print("Total Payable Amount:", units*5)

elif units >= 101 and units <= 200:
    print("Total Payable Amount:", (100*5) + ((units-100)*7))

else:
    print("Total Payable Amount:", (100*5) + (100*7) + ((units-200)*10))
```
